# 해시태그 EDA
- 전체 관광지의 해시태그를 하나의 리스트에 넣어주고 중복제거해서 해시태그 unique 값 확인

해시태그를 하나의 리스트에 넣어주는 코드

```python
# res : range 범위의 Tag_list를 하나의 리스트에 담기 위한 빈 리스트

def tag_join(tag_list, a, b):
    res = []

    for i in range(a, b): # range(a,b)범위의 tag_list를 res 리스트 안에 넣어준다
        res = res + tag_list.iloc[i]

    return res
```

```python
result = tag_join(df_tour['Tag_list'], 0, 354)
```

# 한국어 전처리
- 해시태그를 그냥 쓰기에는 각 해시태그 요소들의 형태가 너무 제각기
- 따라서 해시태그의 형태를 명사 단어로 정제

과정
- 카테고리가 '관광'인 관광지의 모든 해시태그 가져옴
- 관광지별로 해시태그 단어 전처리 > 해시태그 문자열로 바꾸고 명사만 추출
- 관광지별 전처리된 해시태그를 여행 유형별로 분류
- 분류된 단어를 기준으로 관광지별 어떤 유형의 단어가 많이 나오는지 빈도수 확인
- 가장 빈도수가 많은 단어 유형을 해당 관광지의 유형으로 지정

```python
import json
import re
from konlpy.tag import Okt
from collections import Counter
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
from wordcloud import WordCloud
from functools import reduce
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
# 위에서 구한 중복제거되지 않은 관광지 해시태그 모음 리스트
result
```

```python
# 길이로 중복 제거 되지 않은 결과물인 것 확인용
len(result)
```

```python
# 태그 리스트 하나의 문자열로 만드는 코드
# 명사만 정제 시키키 위해서는 문자열의 모음인 리스트 형태면 안됨
# 따라서 각 해시태그 요소를 하나의 문자열에 ' ' 공백을 기준으로 넣어준다.

result_sum = ''
for i in range(len(result)):
    result_sum = result_sum + result[i] + ' '
    # 공백을 기준으로 해시태그들을 하나의 문자열에 집어 넣어줌

result_sum
# 중복 제거되지 않은 전체 해시태그 모음 리스트를 하나의 문장으로 만들어준 결과
```

하나의 문자열로 합쳐진 해시 태그 문자열을 명사만 정제하기

```python
# message_N : result_sum에서 명사만 추출한 명사 리스트

nlp = Okt()
message_N = nlp.nouns(result_sum)
print(message_N)
```

```python
# 전체에서 단어들이 몇번 나왔는지 세어줌
count = Counter(message_N)
count
```

```python
# 상위 빈출 단어 10개 구하는 코드
count.most_common(10)
```

빈출 키워드 자르는 코드

```python
word_count = dict()

# 빈출 단어 50위까지 word_count에 넣어주는 코드
for tag, counts in count.most_common(50):
    if (len(str(tag))>1):
        word_count[tag] = counts
        print(f'{tag} : {counts}')
```

해시태그 빈출 단어(명사) 시각화

```python
font_path = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname = font_path).get_name()
matplotlib.rc('font', family = font_name)
```

```python
plt.figure(figsize = (20, 6))
plt.xlabel('키워드')
plt.ylabel('빈도수')
# plt.grid(True)

sorted_Keys = sorted(word_count, key = word_count.get, reverse = True)
sorted_Values = sorted(word_count.values(), reverse = True)

plt.bar(range(len(word_count)), sorted_Values, align = 'center')

plt.xticks(range(len(word_count)), list(sorted_Keys), rotation = 80)
plt.show()
```

관광지별로 태그 하나의 문자열로 합치고 명사 추출 처리해주기
- 관광지별로 빈도수를 분석할 수 있게 해시태그를 전처리 해줌
- 위에서 진행한 해시태그 명사 단어 추출 전처리 과정과 같으나 위는 전체 해시태그에 대해서 진행했다면 이번 과정은 관광지별로 진행

```python
# 태그 리스트 하나의 문자열로 만드는 코드
result_sum_list = []

for i in range(len(df_tour['Tag_list'])):
    result_sum = ''

    for j in range(len(df_tour['Tag_list'][i])):
        result_sum = result_sum + df_tour['Tag_list'][i][j] + ' '

    result_sum_list.append(result_sum)

result_sum_list
```

```python
# 해시태그를 하나의 문자열(문장)로 합치고 넣을 새로운 컬럼 'tag_word' 생성 후 0 값 집어넣기
df_tour['tag_word'] = 0
```

```python
df_tour[:3]
```

```python
# result_sum_list에서 인덱싱을 사용하여 관광지별로 문장 할당
# result_sum_list[0] 요소가 0번째 관광지의 tag_word 요소가 된다.

for i in range(len(result_sum_list)):
    df_tour['tag_word'][i] = result_sum_list[i]

df_tour['tag_word']
```

```python
# 관광지별로 할당된 문장에서 명사 단어만 추출하는 코드

message_N_list = []
for i in range(len(df_tour['tag_word'])):
    nlp = Okt()
    message_N = nlp.nouns(df_tour['tag_word'][i])
    message_N = list(set(message_N))
    message_N_list.append(message_N)

message_N_list
```

관광지별 분류 속성 카운트 해주는 코드

```python
# 관광지별 해시태그 명사 단어가 해당 단어 분류 리스트에 있는지 확인하는 함수
def is_string_in_list(string, lst):
    return string in lst
```

```python
df_tour_temp.loc[1,'nature']
```

```python
# 관광지별 tag_word_N 리스트에 담긴 명사 단어가 어떤 유형에 속하는지 카운트 해주는 코드

for i in range(len(df_tour_temp['tag_word_N'])):
    nature_count = 0
    history_count = 0
    experience_count = 0
    theme_count = 0
    culture_count = 0
    shop_etc_count = 0
    for j in range(len(df_tour_temp['tag_word_N'][i])):
        if is_string_in_list(df_tour_temp['tag_word_N'][i][j], nature_type):
            nature_count += 1
            df_tour_temp.loc[i,'nature'] = nature_count

        elif is_string_in_list(df_tour_temp['tag_word_N'][i][j], history_type):
            history_count += 1
            df_tour_temp.loc[i,'history'] = history_count

        elif is_string_in_list(df_tour_temp['tag_word_N'][i][j], experience_type):
            experience_count += 1
            df_tour_temp.loc[i,'experience'] = experience_count

        elif is_string_in_list(df_tour_temp['tag_word_N'][i][j], theme_type):
            theme_count += 1
            df_tour_temp.loc[i,'theme'] = theme_count

        elif is_string_in_list(df_tour_temp['tag_word_N'][i][j], culture_type):
            culture_count += 1
            df_tour_temp.loc[i,'culture'] = culture_count

        elif is_string_in_list(df_tour_temp['tag_word_N'][i][j], shop_etc_type):
            shop_etc_count += 1
            df_tour_temp.loc[i,'shop_etc'] = shop_etc_count
```
