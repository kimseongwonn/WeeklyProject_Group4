```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd
from urllib.parse import quote
```

```python
path = '../data/네이버_1601-2100_성원.csv'
df = pd.read_csv(path)
```

# 네이버 지도 추가 정보 크롤링 코드
- 관광지 주소, 관광지별 리뷰 작성 참여자 수, 관광지 분류 속성명

```python
# 웹 드라이버 설정
driver = webdriver.Chrome()
driver.maximize_window()

# @데이터 추출 함수 선언부@

# 추출한 데이터 저장할 저장소 초기화
spot_list = [] # 관광지명 리스트
adress_list = [] # 관광지 주소 리스트
num_person_list = [] # 관광지 리뷰 참여 인원 리스트
spot_attribute_list = [] # 관광지 분류 속성 리스트

#--------------------------------------------------------------------------------------------------------------------
# 크롤링 코드 작성부

# @창 띄우고 관광지 이름으로 검색하는 코드@

# 본인 데이터 지정해서 spots 설정 변경해주세요

# a번
spots = list(temp)[:2]

for j in range(len(spots)):
    spot = spots[j]
    # 특수 문자 포함 관광지명 인코딩해주는 코드
    encoded_spot = quote(spot)

    # 지도 페이지로 바로 들어가도록 변경
    # 기존에는 네이버 검색창에 검색 후 지도 페이지로 연결되도록 2 단계를 거침
    driver.get(f"https://m.map.naver.com/search2/search.naver?query={encoded_spot}")
    time.sleep(5)

#---------------------------------------------------------------------------------------------------------------------
    # 검색 참에서 주소 추출 코드

    # b번

    # button : 주소 정보 펼치는 버튼
    button = WebDriverWait(driver, 5).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="ct"]/div[2]/ul/li[1]/div[1]/div[1]/div/a/i')))
    button.click()

    # adress : 관광지 주소
    adress = driver.find_element(By.XPATH, '//*[@id="ct"]/div[2]/ul/li[1]/div[1]/div[2]/div/p[1]').text

    # spot_list 에 해당 관광지 이름 담고 adress_list에 추출된 관광지 주소 담기
    spot_list.append(spot)
    adress_list.append(adress)

    # 지도창으로 들어와서 관광지 이름 버튼 클릭하는 코드

    # c번

    # map_paths : 지도 검색 페이지에서 해당 관광지의 XPATH 모음 리스트
    map_paths = [
        '//*[@id="ct"]/div[2]/ul/li[1]/div[1]/a/div/strong',
        '//*[@id="ct"]/div[2]/ul/li[1]/div[1]/a[2]/div/strong'
    ]

    # map XPATH를 통해 지도창으로 들어가는 버튼 클릭
    for map_path in map_paths:
        place_button = WebDriverWait(driver, 5).until(
                    EC.element_to_be_clickable((By.XPATH, map_path))
                )

        place_button.click()

        time.sleep(2)

        # 지도창에서 관광지 분류 속성 정보 추출
        spot_attribute = driver.find_element(By.CLASS_NAME, 'lnJFt').text

        # 추출된 관광지 분류 속성 정보 spot_attribute_list에 담기
        spot_attribute_list.append(spot_attribute)
        break


    # 해당 관광지 창으로 들어와서 리뷰버튼 클릭하는 코드 >> 리뷰 참여 인원수 파악하기 위함

    # d번

    review_xpaths = [
        '//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[2]/span',
        '//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[3]/span',
        '//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[4]/span',
        '//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[5]/span'
    ]

    timeout = True
    i = 0
    review_found = False # 리뷰 버튼이 클릭되면 나머지 review_xpath는 클릭 생략하도록 설정
    for review_xpath in review_xpaths:
        if review_found:
            break  # 리뷰를 찾았으면 루프를 종료

        try:
            review_button = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, review_xpath))
            )
            review_button.click()

        except TimeoutException:
            if timeout:
                print('리뷰버튼 못참음')
                # 지도가 리뷰버튼 보다 앞에 존재하는 경우 리뷰버튼으로 넘어가지 못하는 오류 발생
                # 해당 오류 예외처리하고 '시간초과' 문자를 참여 인원 리스트에 삽입하여 리스트 길이 맞춰줌
                num_person_list.append('시간초과')
                timeout = False
                continue

        try:
            time.sleep(2)

            num_person = driver.find_element(By.CLASS_NAME, 'mholG').text

            num_person_list.append(num_person)
            review_found = True

        except Exception as e:
            # 리뷰 버튼을 찾고도 참여 인원 정보를 크롤링하지 못하는 오류 예외 처리
            # 해당 오류 발생의 경우 '공란' 문자열을 삽입 >> 서로 다른 문자열 삽입해서 오류 원인 판단
            print(f"{spot} : 리뷰 버튼이 아닙니다.")
            i += 1

            # 위의 예외 오류 발생시 매번 '공란' 문자열이 추가되는 것을 방지하기 위한 코드
            # 매번 추가하는 것이 아닌 리뷰 버튼 XPATH 최대 경우의 수 4회까지 반복 후 없으면 append
            if i == 4:
                num_person_list.append('공란')
                print('참여인원 오류')
                # continue

# 드라이버 종료
driver.quit()
```

```python
# 리스트 내용 확인 코드
print(spot_list)
print()
print(adress_list)
print()
print(num_person_list)
print()
print(spot_attribute_list)
```

```python
# 크롤링 리스트 4개를 사용하여 데이터 프레임 생성
data = {
    '관광지명': spot_list,
    '주소' : adress_list,
    '리뷰참여인원': num_person_list,
    '관광지분류명': spot_attribute_list
}
df_temp = pd.DataFrame(data)
```
