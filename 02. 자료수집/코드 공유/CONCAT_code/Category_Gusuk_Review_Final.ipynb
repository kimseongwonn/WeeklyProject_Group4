```python
import pandas as pd
```

```python
raw_1 = pd.read_csv('Category_subclass_v2.csv', encoding='cp949')
raw_2 = pd.read_csv('Gusuk_Gusuk_Final_v2.csv', encoding='cp949')
raw_3 = pd.read_csv('Review_Final_v2.csv', encoding='cp949')
```

```python
df1 = raw_1.copy()
df2 = raw_2.copy()
df3 = raw_3.copy()
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
df1.set_index('Attraction', inplace=True)
df2.set_index('Attraction', inplace=True)
df3.set_index('Attraction', inplace=True)

# 세 데이터프레임을 인덱스를 기준으로 결합
concatenated_df = pd.concat([df1, df2, df3], axis=1).reset_index()
```

```python
condition_1 = ['Attraction', 'Category', 'Subclass', 'Address', 'Tag', 'Participants']
Concat_Final = concatenated_df[condition_1]
```

```python
Concat_Final.isnull().sum()
```

```python
Concat_Final.to_csv('Category_Gusuk_Review_Final.csv', encoding='cp949')
```
