```python
Gusuk_1_50 = pd.read_csv('구석구석dataset_1-50_태영.csv',encoding='cp949')
Gusuk_51_100 = pd.read_csv('구석구석dataset_51-100_지우.csv',encoding='cp949')
Gusuk_101_600 = pd.read_csv('구석구석dataset_101-600_수린.csv',encoding='utf-8')
Gusuk_601_1100 = pd.read_csv('구석구석dataset_601-1100_지우.csv',encoding='cp949')
Gusuk_1101_1600 = pd.read_csv('구석구석dataset_1101-1600_태영.csv',encoding='cp949')
Gusuk_1601_2100 = pd.read_csv('구석구석dataset_1601-2100_성원.csv',encoding='utf-8')
```

```python
# Attraction, Gu, Tag
def Change_Col_Names(*Variable_Name):
    for k in Variable_Name:
        k.rename(columns={'attraction': 'Attraction',
                        'location': 'Gu',
                        'tag': 'Tag',
                        'tourist_spot':'Attraction',
                        '주소':'Gu'
                        }, inplace=True)
        k.reset_index(drop = True, inplace=True)


Change_Col_Names(Gusuk_1_50,Gusuk_51_100,Gusuk_101_600,Gusuk_601_1100,Gusuk_1101_1600,Gusuk_1601_2100)
```

```python
condition = ['Attraction','Gu','Tag']

Gusuk_1_50 = Gusuk_1_50[condition]
Gusuk_51_100 = Gusuk_51_100[condition]
Gusuk_101_600 = Gusuk_101_600[condition]
Gusuk_601_1100 = Gusuk_601_1100[condition]
Gusuk_1101_1600 = Gusuk_1101_1600[condition]
Gusuk_1601_2100 = Gusuk_1601_2100[condition]
```

```python
dataframes = [Gusuk_1_50, Gusuk_51_100, Gusuk_101_600, Gusuk_601_1100, Gusuk_1101_1600, Gusuk_1601_2100]
Gusuk_Data_2100 = pd.concat(dataframes, ignore_index=False)
```
