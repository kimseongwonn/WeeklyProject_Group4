## 네이버 기준으로 맞춤

```python
import pandas as pd
```

```python
raw_1 = pd.read_csv('Naver_Data_2100.csv', encoding='cp949')
raw_2 = pd.read_csv('Gusuk_Data_2100.csv', encoding='cp949')
```

```python
df_1 = raw_1.copy()
df_2 = raw_2.copy()
```

## 안맞는거 필터링

```python
condition_1 = pd.DataFrame(df_1['Attraction'])
condition_2 = pd.DataFrame(df_2['Attraction'])
```

```python
condition_1.drop_duplicates(keep='first',inplace=True)
condition_1 = condition_1['Attraction']
condition_1.reset_index(drop=True)
```

```python
dp_2 = condition_2.duplicated()
condition_2[dp_2]
```

```python
condition_2.drop_duplicates(keep='first',inplace=True)
condition_2 = condition_2['Attraction']
condition_2.reset_index(drop=True)
```

```python
merged_df = pd.merge(condition_1, condition_2, on='Attraction', how='outer', indicator=True)
different_values = merged_df[merged_df['_merge'] != 'both']
different_values
```

```python
merged_df['_merge']
```

## 네이버를 구석구석 인기순으로 맞춰주기

```python
Nav = df_1.copy()
Gus = df_2.copy()

# 컬럼이 겹치므로 2로 변경
Gus.columns = ['Attraction2', 'Gu', 'Tag']
Gus
```

```python
# 구석구석 인기순으로 네이버를 맞춰줘야 하므로 구석구석 기준으로 하였음.
result = Gus.merge(Nav, left_on='Attraction2', right_on='Attraction', how='right')
```

```python
# 컬럼 추출
result_1 = result[['Attraction','Good_Point','Selected_People']]
# 인덱스 리셋
result_1.reset_index(drop=True)

Naver_Final = result_1.copy()
```

```python
# 컬럼 추출
result_2 = result[['Attraction2','Gu','Tag']]
# 컬럼 조정
result_2.columns = ['Attraction', 'Gu', 'Tag']

# 구석구석이므로 중복 제거
result_2.drop_duplicates(keep='first', inplace=True)
# 인덱스 리셋
result_2.reset_index(drop=True)

Gusuk_Gusuk_Final = result_2.copy()
```

```python
Naver_Final = result_1.copy()
Gusuk_Gusuk_Final = result_2.copy()
```

```python
Naver_Final.to_csv('Naver_Final.csv', encoding='cp949')
Gusuk_Gusuk_Final.to_csv('Gusuk_Gusuk_Final.csv', encoding='cp949')
```
